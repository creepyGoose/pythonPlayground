**Tuplas Nomeadas (namedtuple)**



```python
import collections
```

```python
coordenadas = collections.namedtuple("Coordenadas", "x y")
p1 = coordenadas(10, 20) # (x,y)
p2 = coordenadas(30, 40)

print(p1, p2)
print(p1.x, p2.y)
```

```python
p1 = p1._replace(x=100) # atualizar valor
print(p1)
```

**Dicionário Padrão**

```python
frutas = ['maçã', 'pêra', 'laranja', 'banana',
         'maçã', 'uva', 'banana', 'banana']

# Use um dicionario default pra contar cada elemento
contador_frutas = collections.defaultdict(int)

# Contando os elementos da lista
for fruta in frutas:
  contador_frutas[fruta] += 1

# Fazendo print do resultado
for (c, v) in contador_frutas.items():
  print(c +": " + str(v))
```

**Contadores**

```python
from collections import Counter
```

```python
turma_a = ["Ezequiel", "Harold", "Yaritza", "Uriel",
    "Dawson","Ezequiel","Karla","Ezequiel",
    "Kinley", "Kinley"]

turma_b = ["Kinley", "Miley", "Jovan", "Karla",
    "Miley", "Kinley", "Abram", "Reginald",
   "Karla", "Miley", "Larry"]

# Criando counter para cada uma das turmas

a = Counter(turma_a)
b = Counter(turma_b)

# Quantos estudantes na turma A se chamam Ezequiel?
print(a["Ezequiel"])

# Quantos alunos existem na turma A?
print(sum(a.values()))
```

```python
# Combine as duas turmas
a.update(turma_b)
print(sum(a.values()))

# Quais são os três nomes mais comuns nas duas turmas?
print(a.most_common(3))
```

```python
# Separe as duas turmas o mostre o nome mais comum 
a.subtract(turma_b)
print(a.most_common(3))
```

```python
# Qual a intersecção entre as duas turmas?
print(a & b)
```

**Dicionário Ordenada**

Mantém a ordem em que são adicionadas

```python
from collections import OrderedDict as o
```

```python
# Lista de times com num. de partidas perdidas e ganhas

times_fut = [("Cruzeiro", (18,12)), ("Vasco", (24, 6)),
             ("Avaí", (20, 10)), ("Flamengo", (22,8)),
             ("Palmeiras", (15, 15)), ("Santos", (20, 10)),
             ("Botafogo", (16, 14)), ("Fluminense", (25, 5))]

# Ordenando os times pela quantidade de vítórias
times_ord = sorted(times_fut, key=lambda t : t[1][0], reverse=True)
```

```python
# Crie um dicionário ordenada com os times
times = o(times_ord)
print(times)
```

```python
# Use popitem para remover o item do topo
nome, est = times.popitem(False)
print("Time mais vitorioso: ", nome, est)
```

```python
# Fazendo um Teste de Igualdade
a = o({"a": 1, "b": 2, "c": 3})
b = o({"a": 1, "b": 2, "c": 3})
print("Teste de igualdade: ", a == b)
```

```python
import collections
import string
```

```python
# Inicialize um deque com letras minusculas
letrinhas = collections.deque(string.ascii_lowercase)

# Ver quantidade de letras
print(len(letrinhas))

# Itere sobre deque
for letra in letrinhas:
  print(letra.upper(), end=',')

# manipule os itens de qualquer um dos terminais
letrinhas.pop()
letrinhas.popleft()
letrinhas.append(2)
letrinhas.appendleft(1)
print(letrinhas)

# Rotacionar a deque
print(letrinhas)
letrinhas.rotate(10)
print(letrinhas)
```
